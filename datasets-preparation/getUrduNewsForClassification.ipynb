```python
def cleaner(article):
    article = re.sub(r'\([^)]*\)', r'', article)
    article = re.sub(r'\[[^\]]*\]', r'', article)
    article = re.sub(r'<[^>]*>', r'', article)
    article = re.sub(r'^https?:\/\/.*[\r\n]*', '', article)
    article = article.replace(u'\ufeff','')
    article = article.replace(u'\xa0', u' ')
    article = article.replace('  ', ' ');
    article = article.replace(' , ', ', ');
    return article
```

```python
all_ent_headlines = []
```

```python
for i in range(1,134):
    url = 'http://urdu.news18.com/movies/page-' + str(i) + '/'
    req = Request(url, headers={'User-Agent': 'Mozilla/5.0'})
    html_doc = ''
    with urlopen(req) as response:
        for line in response:
            line = line.decode('utf-8')
            html_doc = html_doc + line.replace('\n','')
    soup = BeautifulSoup(html_doc, 'html5lib')
    container = soup.find('ul',{'id':'cagetory'})
    lists = container.find_all('li')
    for list_item in lists:
        blog = list_item.find('h2')
        all_ent_headlines.append(cleaner(blog.text))
    print(len(all_ent_headlines))
```

```python
all_ent_headlines = list(filter(lambda x: x != s, all_ent_headlines))
```
